```python
import pandas as pd
```

```python
dataset = pd.read_csv("combinedAirfoilDataLabeled.csv")
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.dropna()
```

```python
y = dataset[["coefficientDrag","coefficientLift"]]
```

```python
x = dataset.drop( columns  = ["coefficientDrag","coefficientLift","airfoilName","botXTR"])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split( x, y, test_size=0.2, random_state = 29)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
x_small = x_train_scaled[:10000]
y_small = y_train[:10000]
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_cd = RandomForestRegressor(n_estimators = 20 , random_state = 29)
model_cd.fit(x_small, y_small)
```

```python
y_test = model_cd.predict(x_test_scaled)
```

```python
print(y_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(y['coefficientDrag'], kde=True)
plt.title('Distribution of Drag Coefficient')
plt.xlabel('coefficientDrag')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.histplot(y['coefficientLift'], kde=True)
plt.title('Distribution of Lift Coefficient')
plt.xlabel('coefficientLift')
plt.ylabel('Frequency')
plt.show()
```

```python
# y_test is a NumPy array — convert to DataFrame for easier plotting
y_test_df = pd.DataFrame(y_test, columns=['coefficientDrag', 'coefficientLift'])

# Plot 1: Actual vs Predicted Drag Coefficient
plt.scatter(y_test_df['coefficientDrag'], y_test[:, 0], alpha=0.5)
plt.xlabel('Actual Drag Coefficient')
plt.ylabel('Predicted Drag Coefficient')
plt.title('Actual vs Predicted Drag Coefficient')
plt.grid(True)
plt.show()

# Plot 2: Actual vs Predicted Lift Coefficient
plt.scatter(y_test_df['coefficientLift'], y_test[:, 1], alpha=0.5)
plt.xlabel('Actual Lift Coefficient')
plt.ylabel('Predicted Lift Coefficient')
plt.title('Actual vs Predicted Lift Coefficient')
plt.grid(True)
plt.show()

```
